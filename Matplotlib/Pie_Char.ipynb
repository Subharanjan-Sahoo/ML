#  Pie Char (Part - 5)

```python
import matplotlib.pyplot as plt
```

```python
classes =["Python", "R" ,"ML" , "AI","DS"]
class1_Students = [45,15,35,25,30]
plt.pie([1])
plt.show
```

```python
classes =["Python", "R" ,"ML" , "AI","DS"]
class1_Students = [45,15,35,25,30]
plt.pie(class1_Students ,labels= classes)
plt.show
```

```python
classes =["Python", "R" ,"ML" , "AI","DS"]
class1_Students = [45,15,35,25,30]
explode=[0,0.3,0.2,0,0]
colors = ["c" ,"r","b","y","g"]
plt.pie(class1_Students ,labels= classes , explode= explode ,colors= colors , autopct= "%0.1f%%",shadow= True, radius =1.4 ,startangle= 270)
plt.show
```
