# Matplotlib Histograms    (Part - 2)

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18,45, (100))
py_student_age =np.random.randint(15,44, (100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,9))
plt.hist(
    [ml_student_age, py_student_age], bins , rwidth= 0.8 ,histtype= "bar", align= "mid", color = ["m","r"] ,label = ["ML Student","Py Student"])

#plt.hist(
 #   py_student_age, bins , rwidth= 0.8 ,histtype= "bar", align= "mid", color = "r" ,label = "Py Student")


plt.title("ML Students age")
plt.xlabel("Students age category")
plt.ylabel("Number of Student age")
plt.legend()
plt.show
```
