# Bar Plot (Part - 3)

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tip_df = sns.load_dataset("tips")
tip_df
```

```python
sns.barplot(x = tip_df.day ,y = tip_df.total_bill)
```

```python
sns.barplot(x = tip_df.day ,y = tip_df.total_bill ,hue=tip_df.sex)
```

```python
sns.barplot(x = 'day' ,y ='total_bill' ,hue='sex', data=tip_df)
```

```python
order = ['Sun','Thur','Fri','Sat']

sns.barplot(x = tip_df.day ,y = tip_df.total_bill ,hue=tip_df.sex,order= order)
```

```python
hue_order = ['Female','Male']

sns.barplot(x = 'day' ,y ='total_bill' ,hue='sex', data=tip_df,hue_order= hue_order)
```

```python
sns.barplot(x = 'day' ,y ='total_bill' ,hue='sex', data=tip_df,hue_order= hue_order, estimator = np.max)
```

```python
sns.barplot(x = 'day' ,y ='total_bill' ,hue='sex', data=tip_df,hue_order= hue_order, estimator = np.max , ci = 20)
```

```python
sns.barplot(x = 'day' ,y ='total_bill' ,hue='sex', data=tip_df,hue_order= hue_order, estimator = np.max, n_boot=2)
```

```python
sns.barplot(y = 'day' ,x ='total_bill' ,hue='sex', data=tip_df,orient= "h")
```

```python
sns.barplot(x= "total_bill", y= 'size',hue = 'sex',data=tip_df,orient='h')
```

```python
sns.barplot(x= "day", y= 'total_bill',hue = 'sex',data=tip_df,palette="hot")
```

```python
sns.barplot(x= "day", y= 'total_bill',hue = 'sex',data=tip_df,dodge=False)
```

```python
sns.set()
kwargs = {'alpha':0.9,'linestyle':':','linewidth':5,'edgecolor':'k'}

sns.barplot(x= "day", y= 'total_bill',data=tip_df,**kwargs)
```

```python
sns.barplot(x= "day", y= 'total_bill',data=tip_df,alpha = .9, linestyle = "-.",linewidth = 6,edgecolor ='g')
```

```python
ax = sns.barplot(x= "day", y= 'total_bill',data=tip_df,alpha = .9, linestyle = "-.",linewidth = 6,edgecolor ='g')

ax.set(title = 'Barplot of Tips DataFrame',xlabel = "Days", ylabel = "Total Bill")
```

```python
plt.figure(figsize= (16,9))

sns.barplot(x= "day", y= 'total_bill',data=tip_df,alpha = .9, linestyle = "-.",linewidth = 6,edgecolor ='g')
plt.title("Barplot of Days and Total Bill")
plt.xlabel("Days")
plt.ylabel("Total Bill")

plt.savefig("Barplot of Days and Total Bill")
plt.show()
```
